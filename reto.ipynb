## 1. Code challenge one
Having a list of n numbers with digits in range [0, S], where n <= 100, switch all list positions in O(n)
time.
If the input number contains a digit greater or equal than S, you will delete the digit from the
number, for example with S=6, 61 becomes 1, and 6 will be deleted from the array. The result
should be printed in console/terminal. Please, don’t use built-in sort of your language.
Examples with S=6:
> [1, 2, 3, 4, 5, 6] -> [5, 4, 3, 2, 1]
> [10, 20, 30, 40] -> [40, 30, 20, 10]
> [6] -> []
> [66] -> []
> [65] -> [5]
> [6, 2, 1] -> [1 , 2]
> [60, 6, 5, 4, 3, 2, 7, 7, 29, 1] -> [1, 2, 2, 3, 4, 5, 0]

```python
S = 6
```

```python
def sort_On(listSort,s):
    if(len(listSort)==1):
        return listSort
    else:
        n=len(listSort)
        for i in range(n // 2):
            listSort[i], listSort[n - 1 - i] = listSort[n - 1 - i], listSort[i]
        return listSort
     
def deleteDigits(listN,s):
    list_New=[]
    for i in listN:
        new_number = ''.join([digit for digit in str(i) if int(digit)<s])
        if new_number:
            list_New.append(int(new_number))
    #print(list_New)
    return sort_On(list_New,s)



list_input = [60, 6, 5, 4, 3, 2, 7, 7, 29, 1]#[61, 23, 45, 17, 96, 52, 64, 77, 89]

if(len(list_input)<=100):
   print( deleteDigits(list_input, S))
else:
    print("Se excede longitud de la lista")
```

## 2. Code challenge two
Write a function that takes in a non-empty array of integers sorted in ascending order and returns a
new array of the same length with the squares of the original integers also sorted in ascending
order. If the output number is out of the range [0, SS] (for S=6 the range will be [0, 66]), you will
delete it of the output array. Please, don’t use built-in sort of your language.
Examples with S=6:
> {"array": [1, 2, 3, 5, 6, 8, 9]} -> [1, 4, 9, 25, 36, 64]
> {"array": [-2, -1]} -> [1, 4]
> {"array": [-6, -5, 0, 5, 6]} -> [0, 25, 25, 36, 36]
> {"array": [-10, 10]} -> []

```python
def sort_Array(list):
    n=len(list)
    for i in range(n):
        for j in range(0, n-i-1):
            if list[j] > list[j+1]:
                list[j], list[j+1] = list[j+1], list[j]
    return list
def squares(arrayInt,s):
     square_list = [x**2 for x in arrayInt if x**2 <= int(str(s)+str(s))]
     return sort_Array(square_list) 

array=[1,2,3,5,6,8,9]
squares(array,S)
```

## 3. Code challenge three (cambiarlo para que no sea googleable)
Given an array of positive integers representing the values of coins in your possession, write a
function that returns the minimum amount of change (the minimum sum of money) that you
CANNOT give change. The given coins can have any positive integer value and aren't necessarily
unique (i.e., you can have multiple coins of the same value). You can use built-in sort of your
language.

```python
from itertools import combinations
def findNumber(setNumbers):
    flag=1
    while flag in setNumbers:
        flag+=1
    return flag

def posibles_sumas(lista):
    combinationSum = []
    for r in range(1, len(lista) + 1):
        combinaciones = combinations(lista, r)
        for combinacion in combinaciones:
            sumaCombination = sum(combinacion)
            combinationSum.append(sumaCombination)
        setSort = set(combinationSum)
    return sorted(setSort)

# Ejemplo de uso
lista_enteros = [5, 7, 1, 1, 2, 3, 22]
setNum= posibles_sumas(lista_enteros)
minimunAmount=findNumber(setNum)
print("Minimun amount:",minimunAmount)


```
